```python
import pandas as pd
from numpy import *
import seaborn as sn 
import matplotlib.pyplot as plt
%matplotlib inline
from numpy import cov
```

```python
df = pd.read_csv('BTC-USD.csv')
```

```python
df.head()
```

# Understanding Dataset

```python
df.describe()
```

```python
df.info()
```

# Checking for null values

```python
df.isnull().sum()
```

# Discovering Correlation b/w dataset

```python
c = df.corr(method='pearson')
c
```

```python
plt.figure(figsize=(10,10))
sn.heatmap(c,fmt=".2f",annot=True)
```

# Scatter plot b/w Open and Volume

```python
plt.figure(figsize=(10,10))
sn.scatterplot(x=df.Open,y=df.Volume/10000000000)
plt.xlabel("Price",fontsize=20)
plt.ylabel("Volume * 10^10",fontsize=20)
```

# Removing Outliers 

```python
outlier = df.Volume[df.Volume<60000000000]
```

```python
outlier
```

```python
plt.figure(figsize=(10,10))
sn.scatterplot(x=df.Open,y=outlier/10000000000)
plt.xlabel("Price",fontsize=20)
plt.ylabel("Volume * 10^10",fontsize=20)
```

# Relation b/w Volume and High

```python
sn.relplot(x=df.High,y=df.Volume/10000000000)
plt.xlabel("High",fontsize=15)
plt.ylabel("Volume * 10^10",fontsize=15)
```

# Removing Outlier

```python
sn.relplot(x=df.High,y=outlier/10000000000)
plt.xlabel("High",fontsize=15)
plt.ylabel("Volume * 10^10",fontsize=15)
```

# Scatterplot b/w Low and Volume after outlier removal

```python
plt.figure(figsize=(10,10))
sn.scatterplot(x=df.Low,y=outlier,palette="BuPu")
plt.xlabel("Low",fontsize=20)
plt.ylabel("Volume * 10^10",fontsize=20)
```

# Multiple Realtion b/w Close, Adj Close and, Volume

```python
plt.figure(figsize=(10,10))
sn.relplot(x=df['Adj Close'],y=outlier)
sn.relplot(x=df.Close,y=outlier,color='r')
```

# High has the highest relation with volume consider correlation heatmap

```python
plt.figure(figsize=(10,10))
sn.relplot(x=df.High,y=outlier,data=df, kind="line")
```
